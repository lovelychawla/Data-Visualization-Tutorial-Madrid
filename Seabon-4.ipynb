```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
from collections import Counter
```

```python
df=sns.load_dataset("tips")
```

```python
#Joint Plot
sns.jointplot(x="tip", y="total_bill", data=df, kind="hex")
```

```python
#Joint plot allows to study the relatioship between 2 numeric variables, the central chart display their correlation, it is usually a scatterplot,
#a hexbin plot and a 2D histogram, or a 2D density plot
```

```python
sns.jointplot(x="tip", y="total_bill", data=df, kind="reg")
```

```python
#Pair Plot
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="sex")
```

```python
#Distribution PLot #Checking the nomalised data
sns.distplot(df["tip"])
```

```python
sns.distplot(df["tip"], kde=False, bins=10)
```

```python
#Categorical Plots
#Count plot
sns.countplot(x="sex", data=df)
```

```python
sns.countplot(y="sex", data=df)
```

```python
#Bar Plot
sns.barplot(x="total_bill", y="sex", data=df)
```

```python
sns.barplot(x="sex", y="total_bill", data=df)
```

```python
#Box Plot-it shows outlyer
sns.boxplot(x="smoker", y="total_bill", data=df, palette="husl")
```

```python
sns.boxplot(x="day", y="total_bill", data=df, palette="rainbow")
```

```python
sns.boxplot(data=df, orient="v")
```
