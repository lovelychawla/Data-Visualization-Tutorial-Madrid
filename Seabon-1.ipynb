```python
sns.set(style="dark")
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
#Relationship Plot
sns.relplot(x="total_bill", y="tip", data=tips, height=4, aspect=3)
```

```python
tips.head()
```

```python
tips["smoker"].value_counts()
```

```python
sns.relplot(x="total_bill", y="tip", hue="sex", data=tips, height=4, aspect=4) #Shift+tab to check defination
```

```python
tips["size"].value_counts()
```

```python
sns.relplot(x="total_bill", y="tip", hue="size", data=tips, height=6, aspect=4)
```

```python
sns.relplot(x="total_bill", y="tip", style="size", hue="sex", data=tips, height=5, aspect=2)
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", style="time", size="size", data=tips, height=5, aspect=2)
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", col="sex", data=tips, height=5, aspect=1)
```

```python
#Scatter Plot
sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
plt.figure(figsize=(20,6))
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="smoker", size="size", style="time")
```

```python
#Count plot
sns.countplot(x=tips["sex"])
```

```python
sns.catplot(x="day", y="total_bill", data=tips, height=5, aspect=2.5)
```

```python
sns.catplot(x="day", y="total_bill", data=tips, height=5, aspect=3, hue="sex")
```
