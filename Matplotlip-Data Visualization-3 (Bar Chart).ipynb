```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
from collections import Counter
```

```python
#Data Visualisation (Bar Chart)
plt.style.available
```

```python
data=pd.read_csv("employee.csv")
```

```python
data.head()
```

```python
#Multivariate Plots
plt.rcParams["figure.figsize"]=(10,4)
x=data.pivot_table(index=["Department", "EducationField"], values=["MonthlyRate"], aggfunc="mean")
x.plot(kind="bar", stacked=False)
plt.xticks(rotation=90)
plt.show()
```

```python
data.groupby(["Department", "EducationField"])["MonthlyRate"].mean()
```

```python
x=pd.crosstab(index=data["Department"],
              columns=data["EducationField"],
              values=data["MonthlyRate"],
              aggfunc="mean")
x
```

```python
plt.rcParams["figure.figsize"]=(8,4)
x.plot(kind="bar", stacked=True)
plt.ylabel("Salary")
plt.xticks(rotation=20)
plt.show()
       
```

```python
#Grouped Bar Plot
plt.rcParams["figure.figsize"]=(8,4)
x.plot(kind="bar", stacked=False)
plt.xticks(rotation=20)
plt.show()
```
