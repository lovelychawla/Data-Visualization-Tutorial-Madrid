```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
from collections import Counter
```

```python
data=pd.read_csv("data_10.csv")
data.head()
```

```python
data.shape
```

```python
data.tail()
```

```python
ages=data["Age"]
dev_salaries=data["All_Devs"]
py_salaries=data["Python"]
js_salaries=data["JavaScript"]
```

```python
plt.plot(ages, py_salaries, label="Python")
plt.plot(ages, js_salaries, label="Javascript")
plt.plot(ages, dev_salaries, color="red", linestyle="--", label="All Devs")
plt.legend()
plt.title("Median Salary (USD)")
plt.tight_layout()
plt.show()
```

```python
fig,(ax1, ax2)=plt.subplots(2,1, sharex=True)
#on the same figure
ax1.plot(ages, dev_salaries, color="green", linestyle="--", label="All Devs")
ax2.plot(ages, py_salaries,label="Python")
ax2.plot(ages,js_salaries, label="Javascript")

ax1.legend()
ax1.set_title("Median Salary (USD) by Age")
ax1.set_ylabel("Median_salary (USD)")

ax2.legend()
ax2.set_xlabel("Ages")
ax2.set_ylabel("Median Salaey (USD)")
plt.tight_layout
plt.show()
```
